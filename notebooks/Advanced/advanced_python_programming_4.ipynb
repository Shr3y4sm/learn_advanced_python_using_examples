```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Separate the data into smokers and non-smokers
smokers = tips[tips['smoker'] == 'Yes']
non_smokers = tips[tips['smoker'] == 'No']

# Plot the histograms
plt.figure(figsize=(12, 6))

# Histogram for smokers
plt.subplot(1, 2, 1)
plt.hist(smokers['total_bill'], bins=15, color='blue', alpha=0.7)
plt.title('Histogram of Total Bill for Smokers')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

# Histogram for non-smokers
plt.subplot(1, 2, 2)
plt.hist(non_smokers['total_bill'], bins=15, color='green', alpha=0.7)
plt.title('Histogram of Total Bill for Non-Smokers')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
