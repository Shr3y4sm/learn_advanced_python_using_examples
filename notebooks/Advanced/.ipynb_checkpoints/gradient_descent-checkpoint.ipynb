```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data for demonstration
np.random.seed(42)
X = 20 * np.random.rand(100, 1)  # Hours studied between 0 and 10
y = 30 + 3.5* X + np.random.randn(100, 1) * 5  # Marks obtained

# Scale the data
X_mean = np.mean(X)
X_std = np.std(X)
X_scaled = (X - X_mean) / X_std

# Hyperparameters
learning_rate = 0.001  # Reduced learning rate
n_iterations = 5000

# Initialize parameters
m = np.random.randn(1)
b = np.random.randn(1)

# Gradient Descent
for i in range(n_iterations):
    # Make predictions
    y_pred = m * X_scaled + b

    # Compute the cost (Mean Squared Error)
    cost = (1 / len(X)) * np.sum((y_pred - y) ** 2)

    # Compute the gradients
    m_grad = (2 / len(X)) * np.sum(X_scaled * (y_pred - y))
    b_grad = (2 / len(X)) * np.sum(y_pred - y)

    # Update parameters
    m -= learning_rate * m_grad
    b -= learning_rate * b_grad

    if i % 100 == 0:  # Print the cost every 100 iterations
        print(f"Iteration {i}: Cost = {cost}, m = {m}, b = {b}")

print(f"Final parameters: m = {m}, b = {b}")

# Plot the data and the linear regression line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, m * X_scaled + b, color='red', label='Regression Line')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.title("Hours Studied vs Marks Obtained")
plt.legend()
plt.show()
```
